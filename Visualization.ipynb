```python
import pandas as pd
import seaborn as sns
import numpy as np
import decimal
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

data = pd.read_excel("Data processing/feature_extracted.xlsx")
```

```python
def float_range(start, stop, step):
  while start < stop:
    yield float(start)
    start += decimal.Decimal(step)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(13,13))

sns.kdeplot(data=data,x="Max Speed(m/s)",ax = axs[0,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="95% Speed(m/s)",ax = axs[0,1],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="75% Speed(m/s)",ax = axs[1,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="Mean Speed(m/s)",ax = axs[1,1],shade=True, color='black',lw = 3)

axs[0,0].tick_params(axis='y', labelsize=12)
axs[0,0].tick_params(axis='x', labelsize=12)
axs[0,1].tick_params(axis='y', labelsize=12)
axs[0,1].tick_params(axis='x', labelsize=12)
axs[1,0].tick_params(axis='y', labelsize=12)
axs[1,0].tick_params(axis='x', labelsize=12)
axs[1,1].tick_params(axis='y', labelsize=12)
axs[1,1].tick_params(axis='x', labelsize=12)
            
axs[0,0].set_yticks(list(float_range(0,0.14,0.02)))
axs[0,1].set_yticks(list(float_range(0,0.14,0.02)))
axs[1,0].set_yticks(list(float_range(0,0.14,0.02)))
axs[1,1].set_yticks(list(float_range(0,0.14,0.02)))
            
axs[0,0].set_title("KDE plot - Max Speed(m/s)",fontweight='bold',fontsize=14)
axs[0,1].set_title("KDE plot - 95% Speed(m/s)",fontweight='bold',fontsize=14)
axs[1,0].set_title("KDE plot - 75% Speed(m/s)",fontweight='bold',fontsize=14)
axs[1,1].set_title("KDE plot - Mean Speed(m/s)",fontweight='bold',fontsize=14)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(13,13))
sns.kdeplot(data=data,x="Max Acceleration(m/s^2)",ax = axs[0,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="95% Acceleration(m/s^2)",ax = axs[0,1],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="75% Acceleration(m/s^2)",ax = axs[1,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="Mean Acceleration(m/s^2)",ax = axs[1,1],shade=True, color='black',lw = 3)

axs[0,0].tick_params(axis='y', labelsize=12)
axs[0,0].tick_params(axis='x', labelsize=12)
axs[0,1].tick_params(axis='y', labelsize=12)
axs[0,1].tick_params(axis='x', labelsize=12)
axs[1,0].tick_params(axis='y', labelsize=12)
axs[1,0].tick_params(axis='x', labelsize=12)
axs[1,1].tick_params(axis='y', labelsize=12)
axs[1,1].tick_params(axis='x', labelsize=12)

axs[0,0].set_yticks(list(range(0,25,3)))
axs[0,1].set_yticks(list(range(0,25,3)))
axs[1,0].set_yticks(list(range(0,25,3)))
axs[1,1].set_yticks(list(range(0,25,3)))

axs[0,0].set_title("KDE plot - Max Acceleration(m/s^2)",fontweight='bold',fontsize=14)
axs[0,1].set_title("KDE plot - 95% Acceleration(m/s^2)",fontweight='bold',fontsize=14)
axs[1,0].set_title("KDE plot - 75% Acceleration(m/s^2)",fontweight='bold',fontsize=14)
axs[1,1].set_title("KDE plot - Mean Acceleration(m/s^2)",fontweight='bold',fontsize=14)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(15,15))
sns.kdeplot(data=data,x="Speed Std",ax = axs[0,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="Acceleration Std",ax = axs[0,1],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="Total Time(s)",ax = axs[1,0],shade=True, color='black',lw = 3)
sns.kdeplot(data=data,x="Total Distance(m)",ax = axs[1,1],shade=True, color='black',lw = 3)

axs[0,0].tick_params(axis='y', labelsize=12)
axs[0,0].tick_params(axis='x', labelsize=12)
axs[0,1].tick_params(axis='y', labelsize=12)
axs[0,1].tick_params(axis='x', labelsize=12)
axs[1,0].tick_params(axis='y', labelsize=12)
axs[1,0].tick_params(axis='x', labelsize=12)
axs[1,1].tick_params(axis='y', labelsize=12)
axs[1,1].tick_params(axis='x', labelsize=12)

axs[0,0].set_title("KDE plot - Speed Std",fontweight='bold',fontsize=14)
axs[0,1].set_title("KDE plot - Acceleration Std",fontweight='bold',fontsize=14)
axs[1,0].set_title("KDE plot - Total Time(s)",fontweight='bold',fontsize=14)
axs[1,1].set_title("KDE plot - Total Distance(m)",fontweight='bold',fontsize=14)

plt.show()
```

```python
data_pair = data[['Transportation Mode', 'Max Speed(m/s)','Max Acceleration(m/s^2)','Total Time(s)','Total Distance(m)']]
fig, axs = plt.subplots(1,2,figsize=(20,10))

sns.scatterplot(data = data_pair,y="Max Speed(m/s)",x="Max Acceleration(m/s^2)",hue='Transportation Mode',ax=axs[0])
sns.scatterplot(data = data_pair,x="Total Time(s)",y="Total Distance(m)",hue='Transportation Mode',ax=axs[1])

sns.set_theme(style='white')

axs[0].get_legend().remove()
axs[1].get_legend().remove()

axs[0].tick_params(axis='y', labelsize=17)
axs[1].tick_params(axis='y', labelsize=17)
axs[0].tick_params(axis='x', labelsize=17)
axs[1].tick_params(axis='x', labelsize=17)

axs[0].set_xlabel('Max Acceleration(m/s^2)',fontsize=17)
axs[0].set_ylabel('Max Speed(m/s)',fontsize=17)
axs[1].set_xlabel('Total Time(s)',fontsize=17)
axs[1].set_ylabel('Total Distance(m)',fontsize=17)

axs[0].set_title("Max speed vs Max acceleration",fontsize = 19,fontweight='bold')
axs[1].set_title("Total distance vs Total time ",fontsize = 19,fontweight='bold')

handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper left', ncol=1,bbox_to_anchor=(.90, 0.70),fontsize=16)

plt.show()
```
